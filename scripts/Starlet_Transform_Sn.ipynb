# Starlet transform on the n-sphere

```python
import sys
sys.path.append("../install/lib/python2.7/site-packages") 
import pyRestoreManifold as prm
```

```python
import numpy as np
import matplotlib.pyplot as plt
from pyRestoreManifold.utils import MakeSignal as mss
import matplotlib

%matplotlib inline
matplotlib.rcParams['figure.figsize']=[20,10]
plt.rcParams['image.cmap']='jet' #'viridis' before
```

## Generate a toy example on S5

```python
t=128
n=5
J=3
tol = 1e-12

X,Ang,x0,v0=mss.Make_Sn(amp=5.,t=t,n=n,w=2)

f, axarr = plt.subplots(2, 2)
axarr[0, 0].imshow(X[0,:,:])
axarr[0, 0].set_title('First observation')
axarr[0, 1].imshow(X[1,:,:])
axarr[0, 1].set_title('Second observation')
axarr[1, 0].imshow(X[2,:,:])
axarr[1, 0].set_title('Third observation')
axarr[1, 1].imshow(X[3,:,:])
axarr[1, 1].set_title('Fourth observation')
```

## Perform the 2D starlet on Sn

```python
# Apply the transform with 4 scales
J = 4
W = prm.trans.ManiStarletSn.Forward2D_Sn(X,J=J,h=[0.0625,0.25,0.375,0.25,0.0625])

```

## Display the coarse scale

The out coefficients have the following size: $(n+1) \times N_x \times N_y  \times (J + 1)$  <br />
They are stored as follows:  <br />
* the local tangent space vectors are stored the first block of size $n \times N_x \times N_y  \times J$
* the decomposition coefficients are stored in the last row-block of size $N_x \times N_y  \times J$
* the coarse scale is stored in the last column-block of size $n \times N_x \times N_y$


```python
f, axarr = plt.subplots(2, 2)
axarr[0, 0].imshow(W[0,:,:,J])
axarr[0, 0].set_title('First observation coarse scale')
axarr[0, 1].imshow(W[1,:,:,J])
axarr[0, 1].set_title('Second observation coarse scale')
axarr[1, 0].imshow(W[2,:,:,J])
axarr[1, 0].set_title('Third observation coarse scale')
axarr[1, 1].imshow(W[3,:,:,J])
axarr[1, 1].set_title('Fourth observation coarse scale')
```

## Display the decomposition coefficients

```python
f, axarr = plt.subplots(2, 2)
axarr[0, 0].imshow(W[n,:,:,0])
axarr[0, 0].set_title('First scale')
axarr[0, 1].imshow(W[n,:,:,1])
axarr[0, 1].set_title('Second scale')
axarr[1, 0].imshow(W[n,:,:,2])
axarr[1, 0].set_title('Third scale')
axarr[1, 1].imshow(W[n,:,:,3])
axarr[1, 1].set_title('Fourth scale')
```
